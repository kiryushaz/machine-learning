## Автомобили

```python
names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels',
         'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders',
         'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
         'highway-mpg', 'price']

auto_df = pd.read_csv('../data/automobile.csv', names=names)
```

```python
auto_df
```

```python
auto_df.describe()
```

```python
auto_df.info()
```

```python
auto_df = auto_df.replace('?', np.nan)
```

```python
auto_df.isnull().sum()
```

```python
auto_df.dtypes
```

```python
auto_df['normalized-losses'] = auto_df['normalized-losses'].astype('float64')
auto_df['normalized-losses'].fillna(auto_df['normalized-losses'].mean(), inplace=True)
```

```python
auto_df.dtypes
```

```python
auto_df.dropna(inplace=True)
```

```python
auto_df
```

```python
auto_df[['bore', 'stroke']] = auto_df[['bore', 'stroke']].astype('float64')
auto_df[['normalized-losses', 'horsepower', 'peak-rpm', 'price']] = auto_df[
    ['normalized-losses', 'horsepower', 'peak-rpm', 'price']].astype('int64')
```

```python
auto_df.info()
```

```python
auto_df.to_csv('../data/automobile_preprocessed.csv')
```
